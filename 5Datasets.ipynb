## 1. Iris Dataset – Scatter Plot

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Iris dataset
iris = sns.load_dataset("iris")

# Display first five records
print(iris.head())

# Scatter Plot
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    s=100
)

plt.title("Scatter Plot of Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

## 2. Tips Dataset – Box Plot

```python
# Load Tips dataset
tips = sns.load_dataset("tips")

# Display dataset
print(tips.head())

# Box Plot
plt.figure(figsize=(8,6))

sns.boxplot(
    data=tips,
    x="day",
    y="total_bill",
    palette="Set3"
)

plt.title("Box Plot of Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")

plt.show()
```

## 3. Titanic Dataset – Count Plot

```python
# Load Titanic dataset
titanic = sns.load_dataset("titanic")

# Display dataset
print(titanic.head())

# Count Plot
plt.figure(figsize=(8,6))

sns.countplot(
    data=titanic,
    x="class",
    hue="survived",
    palette="Set2"
)

plt.title("Titanic Survival Count")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.legend(["Not Survived", "Survived"])

plt.show()
```

## 4. Penguins Dataset – Histogram

```python
# Load Penguins dataset
penguins = sns.load_dataset("penguins")

# Display dataset
print(penguins.head())

# Histogram
plt.figure(figsize=(8,6))

sns.histplot(
    data=penguins,
    x="flipper_length_mm",
    kde=True,
    bins=20,
    color="orange"
)

plt.title("Histogram of Penguin Flipper Length")
plt.xlabel("Flipper Length (mm)")
plt.ylabel("Frequency")

plt.show()
```

## 5. Flights Dataset – Heatmap

```python
# Load Flights dataset
flights = sns.load_dataset("flights")

# Display dataset
print(flights.head())

# Convert data into pivot table
flight_data = flights.pivot(
    index="month",
    columns="year",
    values="passengers"
)

# Heatmap
plt.figure(figsize=(12,7))

sns.heatmap(
    flight_data,
    annot=True,
    cmap="YlGnBu",
    fmt="d"
)

plt.title("Heatmap of Flight Passengers")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```
