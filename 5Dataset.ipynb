```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Iris Dataset – Scatter Plot

```python
# Load Iris dataset
iris = sns.load_dataset("iris")

# Display first five records
print(iris.head())

# Scatter Plot
plt.figure(figsize=(8,6))

sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species",
    s=100
)

plt.title("Scatter Plot of Iris Dataset")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

## Iris Dataset- Pair Plot

```python
import seaborn as sns

iris = sns.load_dataset("iris")

sns.pairplot(
    iris,
    hue="species"
)

plt.show()
```

## Iris Dataset- Line Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

plt.figure(figsize=(8,6))

sns.lineplot(
    data=iris,
    x="sepal_length",
    y="petal_length"
)

plt.title("Line Plot - Iris Dataset")
plt.show()
```

## Iris Dataset - Joint Plot

```python
import seaborn as sns

iris = sns.load_dataset("iris")

sns.jointplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    kind="scatter"
)

plt.show()
```

## 3. Titanic Dataset – Count Plot

## 5. Flights Dataset – Heatmap

```python
# Load Flights dataset
flights = sns.load_dataset("flights")

# Display dataset
print(flights.head())

# Convert data into pivot table
flight_data = flights.pivot(
    index="month",
    columns="year",
    values="passengers"
)

# Heatmap
plt.figure(figsize=(12,7))

sns.heatmap(
    flight_data,
    annot=True,
    cmap="YlGnBu",
    fmt="d"
)

plt.title("Heatmap of Flight Passengers")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```
